### Chart Types and Reasons to Use

Choosing the chart type is predominantly determined by the questions about your data and how to best represent your data for the intended audience.

In addition to picking the proper chart type, there is one other aspect of reporting that is important to remember, scale, i.e., the value you mark on the axis to show the relationship between the units that are being measured.

Often, we may want to utilize multiple chart types to showcase our data.  When doing so, we need to keep scale in mind across our charts, so that our data visualization efforts aren’t misleading.

#### Line Chart
* A line chart is used to identify trends or patterns in data and commonly used for exploring trends over time.
* They can be used to compare multiple groups by using different lines, e.g., the total sales of several products over a period of time.

#### Bar Chart
* A bar chart is most effective when comparing categories of data.
* They can also be used, like a line chart, for tracking changes over time.  When used in this fashion, they are best applied to large changes in the data.
* A bar chart will have two axes, one typically with numerical data and the other with a category, e.g.:
    * The number of different types of apples sold each week at a farmers' market.
    * Or for a time example, the total population of the world since 1000 BCE.


#### Pie Chart
* A pie chart is best used when comparing parts of a whole at a snapshot in time, instead of a change over time.
* A pie chart would answer questions like, what is the percentage of each type of apples sold last week at our farmer's market.  A pie chart will easily show that, for example, 23% were red delicious, 18% were golden delicious, and 15% were granny Smith.

#### Scatter Plots
* Scatter plots are similar to line charts in that they both have a horizontal and vertical axis.
* Scatter plots are used to show how much one variable is impacting another or its correlation.
* We use scatter plots to show relationships between values.  The plot points are markers, and can be different sizes to showcase importance, and different colors to show specific data buckets.
* It allows us to quickly visualize the distribution of the data and notice any outliers.  We can see if there's a positive, negative, or nonexistent correlation between data based on the scatter plot results.

#### Histograms
* Histograms look like bar charts, however, looks can be deceiving as they are not the same, and are indeed used for different purposes.
* Histograms are used to show distributions of variables while bar charts are used to compare the variables.
* Unlike a bar chart a histogram won't have gaps between data. Empty values may be possible however if there are no data points for a particular value.
* In a histogram chart, the data are split into different intervals or bends, to show the frequency of distribution of continuous data. This allows for the inspection of the data distribution, and will show outliers or skewed data.
* When using a histogram, choosing an appropriate number of bins and their width is important for meaningful and accurate reporting.

#### Box Plots
* Box plots, sometimes called box and whisker plots, combine the functionality, the bar chart, with a histogram.
* They allow for the quick examination of and comparison between different sets of data while displaying statistical information about the data.
* It allows for visualizing the minimum first quartile, medium, third quartile, and maximum values of a data set.
* It allows us to see the overall distribution, central value, and variability of a data set.
* Much like a histogram choosing the number of bins and their width is an important consideration for reporting.

#### Heatmap
* A heatmap is a chart in which the area inside recognized boundaries is shaded in proportion to the data being represented.
* For example, you could have a heatmap representing population density.  Countries with higher populations will be represented with different colors than countries with lower populations.

#### Candlestick Charts
* Candlestick charts are heavily used in a financial sector.
* While they bear resemblance to box charts, that where their similarities end.
* Each candlestick will typically show one day of price movement of a stock, currency, or derivative.
* It's like a combination of a line and bar chart showing an overtime trend while also showing the daily information for the data, specifically they show the open, close, high, and low values for security and are a cornerstone of financial technical analysis.

### Our Data Set

Matplotlib relies heavily on NumPy arrays as a data type.

In 1936, a British statistician and biologist named, Ronald Fisher, used a dataset from Edgar Anderson on variations in iris flowers.

This has become a popular dataset to explore and can make for some interesting visualizations.

The dataset takes a look at three species of iris flowers, and the length and width of their flowers, sepals, and petals.

```python
# Read CSV file:
import csv

input_file = 'irises/iris.csv'

with open(input_file, 'r') as iris_data:
    irises = list(csv.reader(iris_data))
```

### Scatter Plot

```python
# Alrady imported matplotlib above
# import matplotlib.pyplot as plt
```

```python
# Color dict
# g is the short code for green
colors = {"Iris-setosa": "#2B5B84", "Iris-versicolor": "g", "Iris-virginica": "purple"}
```

```python
# Get rid of extra item in irises:
irises.pop()
irises
```

```python
# Itertools provides functions for efficient looping:
from itertools import groupby
```

```python
# This often goes under the input file statement (input_file=...), though I'm not sure why
# It needs to go before the other plt calls:
plt.figure(figsize=(12, 8))

for species, group in groupby(irises, lambda i: i[4]):
    categorized_irises = list(group)
    sepal_length = [float(iris[0]) for iris in categorized_irises]
    sepal_width = [float(iris[1]) for iris in categorized_irises]
    plt.scatter(sepal_length, sepal_width, s=10, c=colors[species], label=species)

plt.title("Fisher's Iris Data Set", fontsize=12)
plt.xlabel("sepal length (cm)", fontsize=10)
plt.ylabel("sepal width (cm)", fontsize=10)
plt.legend(loc='upper right')

plt.show()
```

### Histogram
The area of the rectangles in a histogram is proportional to the frequency of the variable.

This allows for the rough assessment of the probable distribution of a given variable.

The rectangles or bins in a histogram, are important to consider when doing data visualization.

Both the number of overall bends and the bend width can have an impact on the overall presentation of data.

```python
virginica_petal_length = []
num_bins = 15 #5 #10

for petal in range(0, len(irises)-1):
    if irises[petal][4] == 'Iris-virginica':
        virginica_petal_length.append(float(irises[petal][2]))
        
plt.figure(figsize=(7.5, 6))
plt.style.use('classic')
plt.hist(virginica_petal_length, num_bins, facecolor='red', alpha=0.75)

plt.title("Iris-virginica Petal Length", fontsize=12)
plt.xlabel("Petal Length (cm)", fontsize=10)
plt.ylabel("Probability", fontsize=10)

# Set the lower and upper bounds
plt.ylim(0, 10)

plt.show()
```

### Box Plots
Each variable shows a box, or rectangle.  The bottom of the rectangle represents the level of the first quartile.

The top of the box represents the third quartile of the data set.

The median is typically represented by a line inside the box.

The minimum and maximum values are represented by whiskers below and above the box respectively.

```python
petal_lengths = []

plt.figure(figsize=(7.5, 4.25))

for species, group in groupby(irises, lambda i: i[4]):
    petal_lengths.append([float(petal[2]) for petal in group])
    
plt.boxplot(petal_lengths)
plt.axis([0, 4, 0, 10])
plt.xticks([1, 2, 3], ['Setosa', 'Versicolor', 'Virginica'])

plt.title("Fisher's Iris Data Set, Petal Length", fontsize=12, loc="left")
plt.xlabel("Iris Variety", fontsize=10)
plt.ylabel("Petal Length (cm)", fontsize=10)

plt.show()
```

### Saving Your Work
